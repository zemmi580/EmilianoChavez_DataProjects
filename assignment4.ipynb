```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2023=pd.read_csv("2023.csv")
df1990=pd.read_csv("1990.csv")
df1995=pd.read_csv("1995.csv")
df2000=pd.read_csv("2000.csv")
df2005=pd.read_csv("2005.csv")
df2010=pd.read_csv("2010.csv")
df2015=pd.read_csv("2015.csv")
df2020=pd.read_csv("2020.csv")

#Ciudad de mexico
#Chihuahua
#Chiapas
dicc = {1990 : df1990, 1995 : df1995, 2000 : df2000, 2005: df2005, 2010 : df2010, 2015 : df2015, 2020 : df2020, 2023 : df2023}
for año, df in dicc.items():
    df['Año'] = año
    df.drop(columns ='Anual')
    
estados = ['Ciudad de México', 'Chihuahua', 'Chiapas']
for fecha,df in dicc.items():
    dicc[fecha]= df[df['Estado'].isin(estados)]
    df['Año'] = fecha
  
#x =pd.merge(df1990,df1995)
df_concatenado = pd.concat(dicc.values(), ignore_index=True)

#df_concatenado.set_index(['Año','Estado'])

#x = df_concatenado.set_index(['Estado','Año']).T


#x['Chiapas']
x = df_concatenado.drop(columns = 'Anual')
x =x.set_index(['Estado','Año']).T
x
```

```python

series_names1 = x['Ciudad de México'].columns
series_names2 = x['Chiapas'].columns
series_names3 = x['Chihuahua'].columns

# Crear un nuevo gráfico
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Seleccionar una paleta de colores
colors = plt.cm.tab20c.colors  # Paleta de colores 'tab10' con 10 colores diferentes
# Iterar sobre cada serie y trazarla en el mismo gráfico con leyenda
#Ciudad de mexico
for i, series_name in enumerate(series_names1):
    ax[0].plot(x['Ciudad de México'].index, x['Ciudad de México'][series_name], label=series_name, color=colors[i])

# Agregar leyendas
ax[0].legend(loc = 'lower center')

# Personalizar etiquetas de ejes y título
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Temperature [C°]')
ax[0].set_title('Mexico city temperature \n over the years')
ax[0].grid(True, alpha=0.1)
for spine in ax[0].spines:
    ax[0].spines[spine].set_visible(False)
    
ax[0].spines['bottom'].set_visible(True)
ax[0].spines['bottom'].set_alpha(0.5)
ax[0].spines['left'].set_visible(True)
ax[0].spines['left'].set_alpha(0.5)
ax[0].set_ylim(10,30)

#Chiapas
for i, series_name in enumerate(series_names2):
    ax[1].plot(x['Chiapas'].index, x['Chiapas'][series_name], label=series_name, color=colors[i])

# Agregar leyendas
ax[1].legend()

# Personalizar etiquetas de ejes y título
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Temperature [C°]')
ax[1].set_title('Chiapas temperature \n over the years')
ax[1].grid(True, alpha=0.1)
for spine in ax[1].spines:
    ax[1].spines[spine].set_visible(False)
    
ax[1].spines['bottom'].set_visible(True)
ax[1].spines['bottom'].set_alpha(0.5)
ax[1].spines['left'].set_visible(True)
ax[1].spines['left'].set_alpha(0.5)
ax[1].set_ylim(10,30)

#Chihuahua
for i, series_name in enumerate(series_names3):
    ax[2].plot(x['Chihuahua'].index, x['Chihuahua'][series_name], label=series_name, color=colors[i])

# Agregar leyendas
ax[2].legend()

# Personalizar etiquetas de ejes y título
ax[2].set_xlabel('Month')
ax[2].set_ylabel('Temperature [C°]')
ax[2].set_title('Chihuahua temperature \n over the years')
ax[2].grid(True, alpha=0.1)
ax[2].set_ylim(10,30)
for spine in ax[2].spines:
    ax[2].spines[spine].set_visible(False)
    
ax[2].spines['bottom'].set_visible(True)
ax[2].spines['bottom'].set_alpha(0.5)
ax[2].spines['left'].set_visible(True)
ax[2].spines['left'].set_alpha(0.5)
fig.suptitle('*Data from every 5 years from 1990', y = 0.01, x =  0.22, alpha = 0.5, fontsize=10)
plt.show()
```
