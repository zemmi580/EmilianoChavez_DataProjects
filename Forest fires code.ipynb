```python
#Import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#read de file
df = pd.read_csv('inc_for.csv')
df
```

```python
# PUT THE DATA IN THE CORRECT FORMAT

df['2014'] = df['2014'].astype(str)
df['2015'] = df['2015'].astype(str)
```

```python
# PUT THE DATA IN THE CORRECT FORMAT
columnas_a_transformar = ['1991', '1995', '2000', '2005',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022', '2023']

for columna in columnas_a_transformar:
        df[columna] = df[columna].str.replace(',', '').astype(int)

df
```

```python
#PUT NAMES OF THE STATES THE SAME TO FURTHER MERGE
df['Estado'][8] = 'Ciudad de México'
df['Estado'][29] = 'Veracruz'
df['Estado'][14] = 'Estado de México'
df['Estado'][4] = 'Coahuila'
df['Estado'][15] = 'Michoacán'
df
```

```python
#SELECT ONLY THE DATA TO ANALYZE
df1 = df[['Estado', '1991', '2023']]
df1
```

```python
#ONLY SELECT THE DATA WE ARE GOING TO ANALYZE
x = ['1991', '2023']
y = [sum(df1['1991']), sum(df1['2023'])]
```

```python
style.use('ggplot')
plt.subplots(figsize=(10, 6))
ax1 = plt.gca() 
plt.bar(x, y, color = 'sandybrown', alpha = 1, width = 0.5)

#DETAILS
plt.title('Comparation of total forestal fires', x =0.5, y = 1.1)
plt.suptitle(' between the years 1991 and 2023',  x=0.5, y=0.95)
for p in ax1.patches:
    ax1.annotate(int(p.get_height()),(p.get_x() + p.get_width() / 2-0.06, p.get_height()-2000))
ax1.set_ylabel('Total forestal fires' , alpha = 1)

#TEXT
plt.text(0.6, 8500, '*Data obtained from: https://www.gob.mx/semarnat', fontsize=8, alpha = 0.5)

plt.savefig('Comparation of total forestal fires.png')
```

```python
#PUT THE DATA INTO THE CORRECT FORMAT
years = ['1991', '1995', '2000', '2005',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022', '2023'] 
df_year = df[years]
df_year_sum = df_year.sum().astype(int)
df_year_sum= pd.DataFrame(df_year_sum)
df_year_sum
```

```python
# GRAPH 3
style.use('ggplot')
plt.subplots(figsize=(10, 6))
plt.plot(df_year_sum.index, df_year_sum.values)
#DETAILS
plt.title('Total of forestal fires through the years', x =0.5, y = 1.1)
plt.suptitle(' from 1991 to 2023',  x=0.5, y=0.95)
#TEXT
plt.text(10, 12600, '*Data obtained from: https://www.gob.mx/semarnat', fontsize=8, alpha = 0.5)

plt.savefig('Total of forestal fires through the years.png')
```

```python
# READ TABLE FROM THE INTERNET
url = 'https://es.wikipedia.org/wiki/Anexo:Entidades_federativas_de_M%C3%A9xico_por_superficie,_poblaci%C3%B3n_y_densidad#cite_note-Censos-4'
tables = pd.read_html(url )
# SELECT THE CORRECT INFORMATION AN CLEAN THE DATA
df_mex = tables[0]
df_mex.drop(['Unnamed: 8', 'Unnamed: 9' ,
       'Unnamed: 10'], axis = 1, inplace = True)
df_mex.drop([0,1], axis = 0, inplace = True)
col_name = {'º': 'Lugar superficie', 'Unnamed: 1': 'Lugar poblacion', 'Unnamed: 2': 'Lugar Densidad', 'Unnamed: 3' : 'Estado', 'Unnamed: 4' : 'Superficie (Km2)',
       'Unnamed: 5' : 'Superficie %', 'Unnamed: 6': 'Poblacion (2020)', 'Unnamed: 7': 'Densidad 2015(hab./km2)'}
df_mex.rename(columns=col_name, inplace=True)
df_mex.set_index('Estado', inplace = True)
df_mex.rename({'Estado de México': 'México'})
df_mex

```

```python
#Merge the two diferent data 
unido = pd.merge(df, df_mex, how = 'outer', on ='Estado')
unido
```

```python
# Put the columns in a correct format to analyze them
unido['Suma de incendios'] = df[['1991', '1995', '2000', '2005',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022', '2023', '2024']].sum(axis=1)
unido['Poblacion (2020)'] = unido['Poblacion (2020)'].str.replace(' ', '').astype(int)
unido.sort_values(by = 'Poblacion (2020)', inplace = True) # Sort for poulation
```

```python
unido
```

```python
#Graph to see the tendecncy
style.use('ggplot')
plt.subplots(figsize=(15, 6))
plt.plot(unido['Poblacion (2020)'],unido['Suma de incendios'])
plt.title('Total of forestal fires through the years', x =0.5, y = 1.1)
plt.suptitle(' from 1991 to 2023',  x=0.5, y=0.95)
plt.text(10, 12600, '*Data obtained from: https://www.gob.mx/semarnat', fontsize=8, alpha = 0.5)
```

```python
# Graph 4
fig, ax1 = plt.subplots(figsize=(16, 6))
plt.bar(unido['Estado'],unido['Suma de incendios'], width = 0.5)

#Details
plt.xticks(rotation=50)
ax1.set_ylabel('Total of forestal fires', color='black')
ax1.set_xlabel('States ', color='black')
plt.title('Forestal fires and population by states', x =0.5, y = 1.1)
ax1.grid(True, color = 'orange', alpha = 0.2)

#Estra text
plt.text(-1.6, 29000, 'Here we can see a comparation between the population from each state and the total forestal fires \nsince the year of 1991 til 2023', fontsize=10, alpha = 0.7)
plt.text(-1.6, 25900, '*Data obtained from: https://www.gob.mx/semarnat, https://cuentame.inegi.org.mx/default.aspx', fontsize=8, alpha = 0.5)

#Second comparation
ax2 = ax1.twinx()
ax2.plot(unido['Estado'], unido['Poblacion (2020)']/1000000, color='green', marker='o')
#Details
ax2.grid(False)
ax2.set_ylabel('Poblacion en millones(2020) ', color='green', rotation = -90)
ax2.yaxis.set_label_coords(1.04, 0.5)
plt.savefig('Forestal fires and population by states.png')
```
